Q1. What is NumPy, and why is it widely used in Python?
NumPy is a powerful Python library used for numerical computing. It provides support for multi-dimensional arrays and matrices along with a collection of mathematical functions. NumPy is widely used because it is faster and more memory-efficient than Python lists. It is the foundation for many other libraries like Pandas, SciPy, and Matplotlib, making it essential for data science and scientific computing.

Q2. How does broadcasting work in NumPy?
Broadcasting in NumPy allows operations between arrays of different shapes. NumPy automatically expands smaller arrays so that they match the shape of larger arrays during operations. This avoids the need for explicit loops, making code cleaner and faster. Broadcasting works only when array dimensions are compatible according to specific rules.

Q3. What is a Pandas DataFrame?
A Pandas DataFrame is a two-dimensional labeled data structure similar to a table or spreadsheet. It consists of rows and columns, where each column can hold different data types. DataFrames make data manipulation easy, including filtering, grouping, and aggregation. They are widely used in data analysis and data preprocessing tasks.

Q4. Explain the use of the groupby() method in Pandas.
The groupby() method in Pandas is used to group data based on one or more columns. After grouping, aggregate functions like sum, mean, or count can be applied. It is commonly used in data analysis to summarize large datasets and extract meaningful insights based on categories.

Q5. Why is Seaborn preferred for statistical visualizations?
Seaborn is preferred because it provides high-level functions for creating attractive and informative statistical graphics. It works seamlessly with Pandas DataFrames and has built-in themes. Seaborn simplifies complex plots like heatmaps, pair plots, and distribution plots, making statistical visualization easier and more readable.

Q6. What are the differences between NumPy arrays and Python lists?
NumPy arrays are faster and more memory-efficient than Python lists. Arrays support vectorized operations, while lists require loops. NumPy arrays store elements of the same data type, whereas lists can store mixed types. Arrays are mainly used for numerical computations, while lists are general-purpose.

Q7. What is a heatmap, and when should it be used?
A heatmap is a graphical representation of data where values are shown using colors. It is commonly used to visualize correlation matrices or patterns in large datasets. Heatmaps help quickly identify relationships, trends, and outliers in data.

Q8. What does “vectorized operation” mean in NumPy?
Vectorized operations allow mathematical operations to be performed on entire arrays at once without using loops. NumPy executes these operations internally in optimized C code, making them faster. This improves performance and code readability.

Q9. How does Matplotlib differ from Plotly?
Matplotlib is a static plotting library mainly used for simple and customized plots. Plotly is an interactive visualization library that supports zooming, hovering, and dynamic updates. Plotly is often used for dashboards and web-based visualizations.

Q10. What is the significance of hierarchical indexing in Pandas?
Hierarchical indexing allows multiple index levels in Pandas DataFrames. It enables working with higher-dimensional data in a two-dimensional structure. This is useful for complex datasets and advanced data analysis.

Q11. What is the role of Seaborn’s pairplot() function?
The pairplot() function creates pairwise relationships between numerical variables in a dataset. It helps visualize distributions and correlations. Pair plots are useful in exploratory data analysis to understand data patterns quickly.

Q12. What is the purpose of the describe() function in Pandas?
The describe() function provides a statistical summary of numerical data. It includes count, mean, standard deviation, minimum, maximum, and quartiles. This helps quickly understand data distribution and detect anomalies.

Q13. Why is handling missing data important in Pandas?
Handling missing data is important because missing values can lead to incorrect analysis. Pandas provides methods like dropna() and fillna() to handle missing data. Proper handling ensures accurate results and reliable models.

Q14. What are the benefits of using Plotly for data visualization?
Plotly provides interactive and visually appealing charts. It supports multiple chart types and works well with web applications. Interactive features like hover and zoom enhance user experience and data exploration.

Q15. How does NumPy handle multidimensional arrays?
NumPy supports multidimensional arrays using the ndarray object. These arrays can represent matrices or higher-dimensional data. NumPy provides indexing, slicing, and reshaping operations to efficiently manage such arrays.

Q16. What is the role of Bokeh in data visualization?
Bokeh is used to create interactive visualizations for web browsers. It supports large datasets and real-time streaming. Bokeh is suitable for dashboards and interactive data exploration.

Q17. Difference between apply() and map() in Pandas?
map() is used on Series for element-wise operations, while apply() works on both Series and DataFrames. apply() can apply functions row-wise or column-wise, making it more flexible.

Q18. What are some advanced features of NumPy?
Advanced NumPy features include broadcasting, vectorization, linear algebra operations, random number generation, and Fourier transforms. These features make NumPy powerful for scientific and numerical computing.

Q19. How does Pandas simplify time series analysis?
Pandas provides date-time indexing, resampling, rolling windows, and time-based filtering. These features make time series analysis easier and more efficient.

Q20. What is the role of a pivot table in Pandas?
Pivot tables summarize data by grouping and aggregating values. They help analyze data across multiple dimensions and are commonly used in reporting and analytics.

Q21. Why is NumPy slicing faster than Python list slicing?
NumPy slicing creates views instead of copies and is implemented in optimized C code. This makes slicing operations faster and more memory-efficient than Python lists.

Q22. What are common use cases for Seaborn?
Seaborn is commonly used for correlation analysis, distribution visualization, regression plots, and categorical data visualization. It is ideal for statistical data analysis.

```python
# 4
# Correlation matrix and heatmap using Seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
sns.heatmap(data.corr(), annot=True)
plt.show()
```

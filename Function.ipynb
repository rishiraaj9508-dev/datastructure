**1. What is the difference between a function and a method in Python?**
A **function** is a standalone block of reusable code defined using `def` and can be called directly by its name. A **method** is a function that is associated with an object or class and is called using the object or class name. Methods usually operate on object data and commonly use `self` to access instance variables.

---

**2. Explain the concept of function arguments and parameters in Python.**
**Parameters** are variables defined in a functionâ€™s definition, while **arguments** are the actual values passed during a function call. Python supports positional, keyword, default, and variable-length arguments (`*args`, `**kwargs`), providing flexibility in how data is passed to functions.

---

**3. What are the different ways to define and call a function in Python?**
Functions can be defined using `def`, lambda expressions, or nested functions. They can be called directly, through a variable reference, as object methods, or by passing them as arguments to other functions.

---

**4. What is the purpose of the `return` statement in a Python function?**
The `return` statement sends a value back to the caller and ends the function execution. A function may return a single value, multiple values as a tuple, or nothing, in which case it returns `None`.

---

**5. What are iterators in Python and how do they differ from iterables?**
An **iterable** is an object that can be looped over, such as a list or string. An **iterator** is an object that produces values one at a time using the `__next__()` method and maintains its state during iteration.

---

**6. Explain the concept of generators in Python and how they are defined.**
Generators are special functions that produce values one at a time using the `yield` keyword. They pause execution after yielding a value and resume from the same point, making them memory-efficient for large data sequences.

---

**7. What are the advantages of using generators over regular functions?**
Generators consume less memory because they generate values on demand. They improve performance for large datasets, support lazy evaluation, and are ideal for handling infinite or very large sequences.

---

**8. What is a lambda function in Python and when is it typically used?**
A lambda function is an anonymous, one-line function defined using the `lambda` keyword. It is commonly used for short operations, especially with functions like `map()`, `filter()`, and `sorted()`.

---

**9. Explain the purpose and usage of the `map()` function in Python.**
The `map()` function applies a given function to each item of an iterable and returns a map object. It is useful for transforming data efficiently without using explicit loops.

---

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**
`map()` transforms elements of an iterable, `filter()` selects elements based on a condition, and `reduce()` combines all elements into a single value using a cumulative operation.

---

**11. Using pen & paper, explain the internal mechanism of sum operation using `reduce()` for the list [47, 11, 42, 13].**
The `reduce()` function applies a binary operation step by step to reduce a list into a single value. For summing `[47, 11, 42, 13]`, `reduce()` starts with the first two elements. First, it adds 47 and 11 to get 58. Next, it takes this result and adds the next element, 42, producing 100. Then, it adds the final element, 13, resulting in 113. Internally, the process works like this:
Step 1: 47 + 11 = 58
Step 2: 58 + 42 = 100
Step 3: 100 + 13 = 113
Thus, `reduce()` repeatedly applies the addition operation until only one final result remains, which is the sum of all elements.

```python
# ============================================================
# Q1) Write a Python function that takes a list of numbers as
#     input and returns the sum of all even numbers in the list.
# ============================================================
def sum_even(numbers):
    return sum(n for n in numbers if n % 2 == 0)

print(sum_even([1, 2, 3, 4, 6, 7]))
```

```python
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Q4) Write a Python function that checks if a given number
#     is prime or not from 1 to 200.
# ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

primes = [n for n in range(1, 201) if is_prime(n)]
print(primes)
```

```python
# >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
# Q5) Create an iterator class in Python that generates the
#     Fibonacci sequence up to a specified number of terms.
# >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

for num in Fibonacci(7):
    print(num, end=" ")
print()



















```

```python

# ================== QUESTION 6 =============================
# Q6) Write a generator function in Python that yields the
#     powers of 2 up to a given exponent.
# ===========================================================
def powers_of_two(exp):
    for i in range(exp + 1):
        yield 2 ** i

for p in powers_of_two(5):
    print(p, end=" ")
print()
```

```python

# ############### QUESTION 7 ###############################
# Q7) Implement a generator function that reads a file line
#     by line and yields each line as a string.
# ##########################################################
def read_file(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line
```

```python
# ++++++++++++++++++ QUESTION 8 +++++++++++++++++++++++++++
# Q8) Use a lambda function in Python to sort a list of tuples
#     based on the second element of each tuple.
# +++++++++++++++++++++++++++++++++++++++++++++++++++++++++
data = [(1, 3), (4, 1), (2, 5)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)

```

```python
# ------------------ QUESTION 9 ----------------------------
# Q9) Write a Python program that uses map() to convert a list
#     of temperatures from Celsius to Fahrenheit.
# ----------------------------------------------------------
celsius = [0, 20, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
# ========== QUESTION 10 ==================================
# Q10) Create a Python program that uses filter() to remove
#      all the vowels from a given string.
# =========================================================
text = "Python Programming"
result = ''.join(filter(lambda ch: ch.lower() not in 'aeiou', text))
print(result)
```
